```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
from sklearn.datasets import make_swiss_roll
from sklearn.decomposition import PCA
from sklearn.manifold import Isomap
```

# Non Linear Dataset

```python
X, t= make_swiss_roll(n_samples=1000, noise=0.2, random_state=0)
```

```python
X.shape
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=t, cmap=plt.cm.Spectral)
```

# PCA on Non linear Dataset

```python
pca= PCA(n_components=2)
xpca= pca.fit_transform(X)
```

```python
xpca.shape
```

```python
plt.scatter(xpca[:,0], xpca[:,1], c=t, cmap=plt.cm.Spectral)
```

# Isomap on Non Linear Dataset

```python
isomap= Isomap()
x_iso= isomap.fit_transform(X)
```

```python
x_iso.shape
```

```python
plt.scatter(x_iso[:,0], x_iso[:,1], c=t, cmap=plt.cm.Spectral)
```
