```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
from sklearn.datasets import make_swiss_roll
from sklearn.decomposition import PCA
from sklearn.manifold import Isomap
```

# Non Linear Dataset

```python
X, t= make_swiss_roll(n_samples=1000, noise=0.2, random_state=0)
```

```python
X.shape
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=t, cmap=plt.cm.Spectral)
```

# PCA on Non linear Dataset

```python
pca= PCA(n_components=2)
xpca= pca.fit_transform(X)
```

```python
xpca.shape
```

```python
plt.scatter(xpca[:,0], xpca[:,1], c=t, cmap=plt.cm.Spectral)
```

# Isomap on Non Linear Dataset

```python
isomap= Isomap()
x_iso= isomap.fit_transform(X)
```

```python
x_iso.shape
```

```python
plt.scatter(x_iso[:,0], x_iso[:,1], c=t, cmap=plt.cm.Spectral)
```

# Non Linear Dimensionality Reduction on Digit Dataset

```python
from sklearn.datasets import load_digits
digits= load_digits()
```

```python
X, y= digits.data, digits.target
X.shape, y.shape
```

# Isomap

```python
isomap= Isomap(n_neighbors=5, n_components= 2)
x_iso= isomap.fit_transform(X)
```

```python
plt.scatter(x_iso[:,0], x_iso[:,1], c=y, cmap= 'Spectral')
```

# MDS

```python
from sklearn.manifold import MDS
mds= MDS(n_components= 2)
x_mds= mds.fit_transform(X)
plt.scatter(x_mds[:,0], x_mds[:,1], c=y, cmap= 'Spectral')
```

# Locally Linear Embedding (LLE)

```python
from sklearn.manifold import LocallyLinearEmbedding
lle= LocallyLinearEmbedding(n_components=2)
x_lle= lle.fit_transform(X)
plt.scatter(x_lle[:,0], x_lle[:,1], c=y, cmap= 'Spectral')
```

# t-SNE

```python
from sklearn.manifold import TSNE
tsne= TSNE(n_components=2)
x_tsne= tsne.fit_transform(X)
plt.scatter(x_tsne[:,0], x_tsne[:,1], c=y, cmap= 'Spectral')
plt.colorbar()
plt.show()
```
