```python
q1= df['fare'].quantile(0.25)
q3= df['fare'].quantile(0.75)
IQR= q3-q1
l_bound= q1-1.5*IQR
u_bound= q3+1.5*IQR
print(l_bound, u_bound)
df= df[(df['fare']>= l_bound) & (df['fare'] <= u_bound)]
```
