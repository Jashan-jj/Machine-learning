```python
from apyori import apriori
rules= apriori(transactions, min_support=0.003, min_confidence=0.2, min_lift=3, min_length=2, max_lenght=2)
```

```python
results= list(rules)
```

```python
results[0]
```

```python
results[2][0]
```

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
suggestions = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```

```python
suggestions.sort_values(by='Lift')
```
